```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dataset = pd.read_csv("wine.csv").values
fitur = dataset[:,:-1]
label = dataset[:,-1]

print(np.corrcoef(fitur, rowvar=False))

plt.figure(figsize=(16,12))
#original Dta
ax1 = plt.subplot(211)
plt.scatter(fitur[:,0], fitur[:,1], c=label)

#Data Transformed With PCA

from sklearn.decomposition import PCA

pca=PCA(n_components=2)
fitur_pca = pca.fit_transform(fitur)

plt.subplot(212, sharex=ax1, sharey=ax1)
plt.scatter(fitur_pca[:,0], fitur_pca[:,1], c=label )
plt.show()
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dataset = pd.read_csv("iris_train.csv").values
fitur = dataset[:,:-1]
label = dataset[:,-1]

#print(np.corrcoef(fitur, rowvar=False))

plt.figure(figsize=(10,8))

#original Dta
ax1 = plt.subplot(311)
plt.scatter(fitur[:,0], fitur[:,1], c=label)

#PCA PLOT
from sklearn.decomposition import PCA

pca=PCA(n_components=2)
fitur_pca = pca.fit_transform(fitur, label)


plt.subplot(312, sharex=ax1, sharey=ax1)
plt.scatter(fitur_pca[:,0], fitur_pca[:,1], c=label )

#Data Transformed With LDA

from sklearn.discriminant_analysis import LinearDiscriminantAnalysis

lda=LinearDiscriminantAnalysis(n_components=2)
fitur_lda = lda.fit_transform(fitur, label)

#LDA PLOT
plt.subplot(313, sharex=ax1, sharey=ax1)
plt.scatter(fitur_lda[:,0], fitur_lda[:,1], c=label )


plt.show()
```
