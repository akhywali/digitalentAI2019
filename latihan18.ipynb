```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

dataset = pd.read_csv("seeds_train.csv").values
fitur = dataset[:,1:]
label = dataset[:,0]

#Corelasivariance
#print(np.corrcoef(fitur, rowvar=False))

plt.figure(figsize=(10,8))

#original Dta
ax1 = plt.subplot(412)
plt.scatter(fitur[:,0], fitur[:,1], c=label)

#PCA PLOT
from sklearn.decomposition import PCA
pca=PCA(n_components=2)
fitur_pca = pca.fit_transform(fitur, label)
plt.subplot(413, sharex=ax1, sharey=ax1)
plt.scatter(fitur_pca[:,0], fitur_pca[:,1], c=label )

#Data Transformed With LDA
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
lda=LinearDiscriminantAnalysis(n_components=2)
fitur_lda = lda.fit_transform(fitur, label)

#LDA PLOT
plt.subplot(414, sharex=ax1, sharey=ax1)
plt.scatter(fitur_lda[:,0], fitur_lda[:,1], c=label )


plt.show()
```
