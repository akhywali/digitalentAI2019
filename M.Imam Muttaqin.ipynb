```python
def fibo(x):
    fibo_list = []
    a = 0
    b = 1
    for i in range(x):
        fibo_list.append(a)
        if b > x:
            return fibo_list
        a, b = b, a+b
        
num = int(input())
print(max(fibo(num)))
```

```python
n = int(input())

for x in range(0, n):  
    for y in range(0, n): 
        if (x == 0 or x == n-1 or y == 0 or y == n-1 or x == y or x == n-1-y):          
            print( "*", end="")  
        else: 
            print( " ", end="") 
    print("") 
```
