```python
jumlah_vendor=data_read['vendor_name'].value_counts()
vendor_10=jumlah_vendor>10
vendor_to_keep=list(np.array(jumlah_vendor.index.values)[np.array(vendor_10)])
rows_before=data_read.shape[0] #jumlah baris sebelumnya
vendor_baru=data_read.loc[data_read["vendor_name"].isin(vendor_to_keep)]
rows_after=vendor_baru.shape[0] #jumlah baris setelah dihilangkan
print("{} rows removed!".format(rows_before-rows_after))
```

```python
vendor_baru.reset_index(drop=True, inplace=True)
vendor_baru
```
